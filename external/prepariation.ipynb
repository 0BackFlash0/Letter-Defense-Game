```python
import pandas as pd
import os

cur_dir = os.getcwd()

files = os.listdir(cur_dir)

db_list = []
```

```python
cur_dir
```

```python
for name in files:
    _, extension = os.path.splitext(name)

    if(extension==".xls"):
        df = pd.read_excel(cur_dir+"\\"+name)
        db_list.append(df)
```

```python
db_list
```

```python
word_df = pd.concat(db_list, ignore_index=True)
```

```python
word_df.columns
```

```python
# word_df.drop(['원어·어종', '원어', '어원', '발음', '활용', '검색용 이형태', '문형', '문법', '범주', '용례', '관련 어휘', '속담', '관용구', '대역어', '생물 분류군 정보', '역사 정보', '수어 정보', '규범 정보', '다중 매체(멀티미디어) 정보'], axis=1, inplace=True)
```

```python
word_df
```

```python
word_df = word_df[word_df['구성 단위']=="단어"]
```

```python
word_df[word_df['고유어 여부']!="외래어"]
```

```python
word_df['어휘'] = word_df['어휘'].str.replace('-', '', regex=False)
```

```python
word_df['어휘'] = word_df['어휘'].str.replace('^', '', regex=False)
```

```python
word_df = word_df[~(word_df['뜻풀이'].str.contains('의 옛말') | word_df['뜻풀이'].str.contains('의 준말'))]
```

```python
word_df = word_df[(word_df['방언 지역'].isna() & word_df['생물 분류군 정보'].isna() & word_df['역사 정보'].isna() & word_df['수어 정보'].isna() & word_df['전문 분야'].isna())]
```

```python
word_df = word_df[word_df['고유어 여부']=="고유어"]
```

```python
word_df.to_csv("./word_df.csv")
```

```python
word_df[word_df.duplicated(subset='어휘', keep=False)]
```

```python
word_df['의미 번호'] = word_df.groupby('어휘').cumcount()+1
```

```python
word_df
```

```python
result_df = word_df.groupby('어휘').apply(lambda x: '\n'.join([f"{i}. {b}" for i, b in zip(x['의미 번호'], x['뜻풀이'])])).reset_index(name='합쳐진 뜻풀이')
```

```python
result_df
```

```python
result_df = result_df[354:].reset_index()
```

```python
result_df = result_df[~(result_df['어휘'].str.contains('') | result_df['어휘'].str.contains(''))]
```

```python
result_df.reset_index(inplace=True)
```

```python
result_df.drop(['level_0', 'index'], axis=1, inplace=True)
```

```python
result_df.to_csv('./result_df.csv')
```
